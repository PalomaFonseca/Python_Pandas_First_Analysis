# **QUESTÕES E ANÁLISES**

```python
#Qual a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
#Criando a coluna de custo primeiro
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head(1)
```

```python
#Qual o custo total?
#Arredondando em duas casas decimais
round(df["Custo"].sum(), 2)
```

```python
#Com a receita e o custo é possível encontrar o lucro
#Como criar uma coluna de lucro?
df["Lucro"] = df["Valor Venda"] - df["Custo"]
df.head(1)
```

```python
#Qual o valor total do lucro?
round(df["Lucro"].sum(), 2)
```

```python
#Qual o total de dias para envio do produto?
df["Tempo_Envio"] = df["Data Envio"] - df["Data Venda"]
df.head(1)
```

```python
#Qual a média do tempo de envio para cada marca?
#1-Tranformar coluna Tempo_Envio em numerica pois para calcular a média não pode ser com string
#Extraindo apenas os dias
df["Tempo_Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head(1)
```

```python
#Validando o tipo de dado do Tempo_Envio
df["Tempo_Envio"].dtype
```

```python
#Qual a média do tempo de envio por marca?
df.groupby("Marca")["Tempo_Envio"].mean().sort_values(ascending=True) 
```

```python
#Existem valores faltantes na base?
df.isnull().sum()
```

```python
#Qual o lucro por ano e por marca?
#1-Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Retirando notação científica do Pandas
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Deixando em formato de coluna, resetando o index
Lucro_Ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
Lucro_Ano
```

```python
#Qual o total de produtos vendidos?
df["Quantidade"].dtype
df["Quantidade"].sum()
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = True)
```

```python
#Total de produtos vendidos em um gráfico
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False).plot.barh(title = "Total de produtos vendidos", color="gray")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Qual o lucro por ano?
df.head(1)
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Gráfico de lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title = "Lucro por ano", color = "Green")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
#Quais foram as vendas do ano de 2009?
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
#Quais foram os lucros por mês do ano de 2009?
df.groupby(df["Data Venda"].dt.month)["Lucro"].sum()
```

```python
#Gráfico de lucro por mês do ano de 2009
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot.bar(title = "Lucro por mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Gráfico de lucro por marca do ano de 2009
df.groupby(df["Marca"])["Lucro"].sum().plot.bar(title = "Lucro por marca", color = "magenta")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');#Alterando a rotação do nome do eixo x
```

```python
#Gráfico de lucro por classe do ano de 2009
df.groupby("Classe")["Lucro"].sum().plot.bar(title = "Lucro por classe", color = "yellow")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation = 'horizontal');
```
