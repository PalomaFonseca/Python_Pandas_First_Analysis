```python
#Importando a biblioteca pandas apelidando a mesma como pd
import pandas as pd
```

```python
#Lendo o arquivo através de um caminho, ignorando erros de linha e usando ";" como separador pois nesse caso específico não foi usado a "," que é o padrão
df = pd.read_csv("/content/drive/MyDrive/DataSets/Gapminder.csv",error_bad_lines = False, sep=";")
```

```python
#Visualizando as 10 primeiras linhas
df.head(10)
```

```python
#Renomeando as colunas
df = df.rename(columns={"country": "País", "continent": "Continente", "year": "Ano", "lifeExp": "ExpecVida", "pop": "População", "gdpPercap": "PIB"})
```

```python
#Trazendo valores únicos de uma coluna
df["Continente"].unique()
```

```python
#Trazendo valores contidos em Oceania
Oceania = df.loc[df["Continente"] == "Oceania"]
df.head()
Oceania["Continente"].unique()
```

```python
#Agrupando quantidade de países por continente
df.groupby("Continente")["País"].nunique()
```

```python
#Agrupando média expectativa de vida para cada ano
df.groupby("Ano")["ExpecVida"].mean()
```
